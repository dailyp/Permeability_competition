# Extratrees

Extremely Randomized Trees (Extra trees for short) makes trees even more random by also using random thresholds for each feature rather than searching for the best possible thresholds (like regular Decision Trees do)

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_squared_log_error
from sklearn.ensemble import RandomForestRegressor
#from sklearn.preprocessing import StandardScaler, scale  #scaling not necessary for random forest

pd.set_option("display.max_rows", 500)
pd.set_option("display.max_columns", 500)
```

```python
perm_train = pd.read_csv('data/train.csv')
perm_train.head()

```

```python
# start over and do a standard random forest 
# use all columns of the perm_train dataset.  Drop permeability from X, keep only permeability for y

#Define predictors
X = perm_train.drop(columns = ['permeability'])

#Define target
y = perm_train['permeability']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
#note ExtraTreesClassifier is used for classification data, while Regressor is used for numeric
```

```python
et = ExtraTreesRegressor(n_estimators=800, max_features=290, min_samples_split=2).fit(X_train, y_train)
```

```python
mean_absolute_error(y_test, et.predict(X_test))
```

```python
np.sqrt(mean_squared_error(y_test, et.predict(X_test))) 
```

```python
np.sqrt(mean_squared_log_error( y_test, et.predict(X_test)))
```

Not the best RMSLE score that we've seen, but will submit to kaggle anyway.  Apply the model to the test set and convert to preferred competition format.
